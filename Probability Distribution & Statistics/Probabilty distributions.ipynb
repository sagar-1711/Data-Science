# Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generate a normal distribution(1-D)
#Standard normal distribution
u=0
sigma=1
vals=u + sigma*np.random.randn(1000)
print(vals.shape)
print(vals)
```

```python
plt.hist(vals,50) #50 denotes thickness of bar
plt.show()
```

```python
u=30
sigma=10
vals=u + sigma*np.random.randn(1000)
plt.hist(vals,100) #50 denotes thickness of bar or intervals
plt.show()
```

```python
vals=np.round(vals)
z=np.unique(vals,return_counts=True)
print(z)
# max probability is around 30 i.e. 43
```

```python
u=5
sigma=1
vals=u + sigma*np.random.randn(1000)
plt.hist(vals,10) #50 denotes thickness of bar or intervals
plt.show()
```

```python
x=vals
y=np.zeros(x.shape)
plt.scatter(x,y)
plt.show()
#most of the points are around 5
```

# Multivariate Normal Distribution

```python
mean=np.array([0,0])
cov=np.array([[1,0],[0,1]])
dist=np.random.multivariate_normal(mean,cov,500)
print(dist.shape)
print(dist)
```

```python
plt.scatter(dist[:,0],dist[:,1])
plt.show()
```

```python
mean=np.array([0,0])
cov=np.array([[1,0.5],[0.5,1]])
dist=np.random.multivariate_normal(mean,cov,500)
print(dist.shape)
plt.scatter(dist[:,0],dist[:,1])
plt.show()
#axis has slightly tilted towards the line x=y
```

```python
mean=np.array([0,0])
cov=np.array([[1,0.8],[0.8,1]])
dist=np.random.multivariate_normal(mean,cov,500)
print(dist.shape)
plt.scatter(dist[:,0],dist[:,1])
plt.show()
#tilting is more
```

```python
mean1=np.array([0,0])
cov1=np.array([[1,0],[0,3]])#sigmay is more so spread will be more along y axis in dist1 
mean2=np.array([5,6])
cov2=np.array([[1.3,0.2],[0.2,1.1]])

dist1=np.random.multivariate_normal(mean1,cov1,500)
dist2=np.random.multivariate_normal(mean2,cov2,500)

plt.scatter(dist1[:,0],dist1[:,1])
plt.scatter(dist2[:,0],dist2[:,1])

plt.show()
```

```python
mean=np.array([0,0])
cov=np.array([[1,-1],[-1,1]])#negative correlation increase in x will decrease y
dist=np.random.multivariate_normal(mean,cov,500)
print(dist.shape)
plt.scatter(dist[:,0],dist[:,1])
plt.show()
```
