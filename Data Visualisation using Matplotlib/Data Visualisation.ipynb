```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
themes=plt.style.available
print(themes)
```

```python
plt.style.use("seaborn")
```

```python
#Line Plot
# 2 arrays are reqiuired
x=np.arange(10)
y1=x**2
y2=2*x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1)
plt.show()
```

```python
plt.plot(x,y1,color="red")
plt.show()
```

```python
plt.plot(x,y1,color="red",label="Apple",marker='o')
plt.plot(x,y2,color="green",label="kiwi",linestyle="dashed",marker='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of fruits Over Time")
plt.legend()
plt.show()
# if we want to pop out window in python script then we have to write plt.show(), in jupyter it will work fine 
# but in py script we will not get output
```

```python
#Another reason for plt.show()->if we want to display two graphs in same shell
plt.plot(x,y1,color="red")
plt.show()
plt.plot(x,y2,color="green")
plt.show()
```

```python
prices=np.array([1,2,3,4])**2
print(prices)
plt.plot(prices)# prices becomes Y and indices of array becomes X
plt.show()
```

# Scatter Plots 

```python
plt.scatter(x,y1,color="red",label="Apple",marker='^')
plt.scatter(x,y2,color="green",label="kiwi",linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plot-Prices Over Time")
plt.legend()
plt.show()
```

```python
#Adjust the size of any plot
plt.figure(figsize=(4,4))
plt.plot(x,y1,color="red")
plt.show()
```

# Bar Graphs

```python
plt.bar?
```

```python
x=[0,1,2]
y=[10,20,15]
plt.bar(x,y)
plt.show()
```

```python
x_coordinates=np.array([0,1,2])*2
y1_coordinates=[10,20,15]
y2_coordinates=[20,10,12]
plt.bar(x_coordinates,y1_coordinates,width=0.5,label="Current Year",tick_label=["Gold","Platinum","Silver"],color="blue")
plt.bar(x_coordinates+0.5,y2_coordinates,width=0.5,label="Next Year",tick_label=["Gold","Platinum","Silver"],color="orange")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.title("Metal Price Comparison")
plt.xlim(-2,5)
plt.ylim(0,40)
plt.legend()
plt.show()
```

# Pie Chart

```python
weightage=[20,10,15,5]
subjects=["Maths","Chem","Physics","English"]
plt.pie(weightage,labels=subjects)
plt.show()
```

```python
plt.style.use("seaborn")
weightage=[20,10,15,5]
subjects=["Maths","Chem","Physics","English"]
plt.pie(weightage,labels=subjects,explode=(0,0,0.2,0),autopct='%1.1f%%',shadow=True)
plt.title("Subjects")
plt.show()
```

# Standard Normal Distribution

```python
Xsn=np.random.randn(100)
#returns a sample from from standard normal distribution
print(Xsn)
```

```python
sigma=5
u=60
X=Xsn*sigma + u
print(X)
```

```python
X=np.round(X)
print(X)
X2=np.round(Xsn*15+40)
```

# Histogram

```python
plt.style.use("seaborn")
plt.hist(X,alpha=1,label="Maths")
plt.hist(X2,alpha=0.8,label="Physics")# alpha gives opacity
plt.xlabel("Marks Range")
plt.ylabel("Frequncy of Students")
plt.legend()
plt.show()
```
