```python
titles=df.get('movie_title')
print(titles)
```

```python
## let us try to gain some insights
## Plot X-len(Moviename), Y-frequency
freq_title={}

for t in titles:
    length=len(t)
    
    try:
        freq_title[length]+=1
    except:
        freq_title[length]=1

print(freq_title)
```

```python
x=freq_title.keys()
y=freq_title.values()
print(type(x))
print(x)
```

```python
X=list(x)
Y=list(y)
plt.xlabel("Length of Title")
plt.ylabel("No.of movies")
plt.scatter(X,Y)
plt.show()
#somewhat looks like normal distribution
```
